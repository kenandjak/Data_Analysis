Conjunto de dados: as 50 mulheres mais influentes do mundo em 2012

```python
import pandas as pd

ages = [26, 31, 35, 37, 43, 43, 43, 44, 45, 47, 48, 48, 49, 50, 51,
51, 51, 51, 52, 54, 54, 54, 54, 55, 55, 55, 56, 57, 57, 57,
58, 58, 58, 58, 59, 59, 59, 62, 62, 63, 64, 65, 65, 65, 66,
66, 67, 67, 72, 86]
```

### MEASURES OF CENTRAL TENDENCY

1. Finding a sample mean:

```python
# age_table = pd.DataFrame({
#     'mean': ages.mean()
# })
mean = sum(ages) / len(ages)
print(mean)
```

2. Fiding the median:

```python
# median = sorted(ages)[len(ages) // 2]
# print(median)
median = (ages[len(ages) // 2] + ages[(len(ages) - 1) // 2]) / 2
print(median)
```

3. Finding the mode:

```python
count1 = 1
count2 = 0
mode1 = 0
for i in range (len(ages) - 1):
  if ages[i] == ages[i+1]:
    count1 += 1
    mode1 = ages[i]
  else:
    if count2 < count1:
      count2 = count1
      count1 = 1
      mode2 = mode1
if count1 > count2:
  print(mode1)
elif count2 > count1:
  print(mode2)
else:
  print(mode1, mode2)
```

### MEASURES OF VARIATION


1. Finding the desviations:

```python
deviations = []

for i in range (len(ages)):
   #valor = float("{:.2f}".format(ages[i] - mean))
   valor = round(ages[i] - mean,2)
   deviations.append(ages[i] - mean)
```

```python
age_table = pd.DataFrame({
    'age': ages,
    'deviation': deviations
})
print(age_table)
```

2. Finding the variance:

```python
age_table['squared_deviation'] = age_table['deviation'] ** 2
print(age_table)
variance = age_table['squared_deviation'].sum() / 50
print("Population Variance:", variance)
```

3. Finding the standard deviation:

```python
std_dev = variance ** 0.5
print("Population Standard Deviation:", std_dev)
```

### MEASURES OF POSITION

1. Finding the quartiles:

```python
q1 = ages[len(ages) // 4]
q2 = ages[len(ages) // 2]
q3 = ages[3 * len(ages) // 4]
print("First Quartile:", q1)
print("Second Quartile:", q2)
print("Third Quartile:", q3)
```

2. Interquartile Range and outliers:

```python
iqr = q3 - q1
print("Interquartile Range:", iqr)
for i in range (len(ages)):
  if ages[i] < q1 - 1.5 * iqr or ages[i] > q3 + 1.5 * iqr:
    print("Outlier: ", ages[i])
```

3. Drawing a boxplot:

```python
import matplotlib.pyplot as plt
plt.boxplot(ages, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))
plt.title('Boxplot of the 50 most influential women in the world (2012)')
plt.xlabel('Ages')
plt.show()
```
