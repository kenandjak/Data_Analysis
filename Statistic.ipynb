**FREQUENCY DISTRIBUTION**



---



1. Constructing a frequency distribution - defining the data:

```python
import pandas as pd
```

```python
num_classes = 7
# As 50 mulheres mais influentes em 2012:
ages = [26, 31, 35, 37, 43, 43, 43, 44, 45, 47, 48, 48, 49, 50, 51,
51, 51, 51, 52, 54, 54, 54, 54, 55, 55, 55, 56, 57, 57, 57,
58, 58, 58, 58, 59, 59, 59, 62, 62, 63, 64, 65, 65, 65, 66,
66, 67, 67, 72, 86]
```

2. Determine the range of data:

```python
minimum = min(ages)
maximum = max(ages)
range = maximum - minimum
print("Range: ",range)
```

3. Determine the class width:

```python
class_width = round(range / num_classes)
print("Class Width: ", class_width)
```

4. Finding class limits:

```python
# df["col"][row_indexer] = value
# df.loc[row_indexer, "col"] = values
age_table = pd.DataFrame({
    'lower limit':[26],
    'upper limit':[34]
})
low_lim = 26
upp_lim = 34
count = 1
while(count < num_classes):
    low_lim += class_width
    upp_lim += class_width
    age_table.loc[count,'lower limit'] = low_lim
    age_table.loc[count,'upper limit'] = upp_lim
    count += 1
display(age_table)
```

5. Total frequency for each class:

```python
for i, row in age_table.iterrows():
  freq = 0
  for age in ages:
    if (age >= row['lower limit'] and age <= row['upper limit']):
      freq += 1
    age_table.loc[i,'frequency'] = freq
display(age_table)
```

Análise dos dados pela frequência:
- A maioria das mulheres mais influentes estão na 4ª classe (53-61)
- Apenas 4.0% delas têm menos de 35 anos

6. Relative and Cumulative Frequencies:

```python
for i, row, in age_table.iterrows():
  relative_freq = row['frequency']/len(ages)
  age_table.loc[i,'relative frequency'] = relative_freq

cumul_freq = 0
for i, row, in age_table.iterrows():
  cumul_freq = row['frequency'] + cumul_freq
  age_table.loc[i, 'cumulative frequency'] = cumul_freq
display(age_table)
```

7. Histogram:

```python
import matplotlib.pyplot as plt
plt.hist(ages,num_classes)
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Ages of 50 most powerful women (2012)')
plt.show()
```

⭐ Another graph:

```python
age_table.plot(kind="scatter", y="upper limit", x="lower limit", c="frequency", marker='*', s=70)
plt.show()
```
